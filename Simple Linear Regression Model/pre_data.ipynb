```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

Discover The DataSet

```python
df = pd.read_csv("house Prediction Data Set.csv",header=None)
df.head(10)
```

```python
def split_to(row):
    # Split each row by comma and convert to numeric values
    return pd.Series([float(x) for x in row[0].split()])

data = df.apply(split_to, axis=1)
data.head()

```
