```python
import pandas as pd
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score ,confusion_matrix ,recall_score ,precision_score
from sklearn.preprocessing import LabelEncoder , StandardScaler
```

# Discover Data

```python
df = pd.read_csv("iris.csv")
df.head()
```

# Featuer Selection

```python
lb = LabelEncoder()
df["target"] = lb.fit_transform(df["species"])
df.drop("species", axis=1 , inplace=True)
```

```python
df.corr()["target"].sort_values(ascending=False)
```

```python
x = df.drop('target' , axis = 1)
y = df["target"]
```

```python
x_train ,x_test , y_train , y_test = train_test_split(x , y ,test_size=0.3 , random_state=42)
```

```python
scaler = StandardScaler()
x_train_scaled = scaler.fit_transform(x_train[["petal_width" , "petal_length", "sepal_length","sepal_width"]])
x_test_scaled = scaler.transform(x_test[["petal_width" , "petal_length", "sepal_length","sepal_width"]])
 
```

```python
model = KNeighborsClassifier()
```

```python
model.fit(x_train_scaled , y_train)
y_pred = model.predict(x_test_scaled)
```

```python
print("#"*35)
print("confusion_matrix \n",confusion_matrix(y_pred=y_pred , y_true=y_test))
print("#"*35)
print("accuracy_score ",accuracy_score(y_pred=y_pred , y_true=y_test))
print("#"*35)
print("precision_score ",precision_score(y_pred=y_pred , y_true=y_test , average='weighted'))
print("#"*35)
print("recall_score ",recall_score(y_pred=y_pred , y_true=y_test , average='weighted'))
print("#"*35)
```
